## How to Flatten Nested Columns

```python
df.belongs_to_collection[0]
```

```python
# in the belongs_to_collection column, we need only 'name' (the name refers to collection of the name)
```

```python
df.belongs_to_collection = df.belongs_to_collection.apply(lambda x: x['name'] if isinstance(x, dict) else np.nan) 
```

```python
df.belongs_to_collection.value_counts(dropna = False).head(20)
```

```python
df.genres
```

```python
# in genres column we have lists of dictionaries
```

```python
df.genres = df.genres.apply(lambda x: "|".join(i['name'] for i in x))
```

```python
df.genres.value_counts(dropna = False).head(20)
```

```python
# genres column'ında name filmin türünü belirtir. görüldüğü üzere 2442 adet name'i belirtilmemiş film var. 
# bunlardan kurtulmamız için boşlukları np.nan ile doldur
```

```python
df.genres.replace("", np.nan, inplace = True)
```

## Cleaning Numerical Columns(PART 1)

```python
df.info()
```

```python
df.budget
```

```python
# as can be seen, budget column consists of 'object' type of values
```

```python
# let's try to convert object to float
df.budget.astype('float')
```

```python
# there are different type of values in budget column. So we cant simply convert it to float
```

```python
df.budget = pd.to_numeric(df.budget, errors = 'coerce')
```

```python
df.budget.value_counts(dropna = False)
```

```python
# 0 budget cannot exist so we replace 0s with np.nan
df.budget = df.budget.replace(0, np.nan)
```

```python
# to improve the readibility, we divide the money(in budget) 1000000
df.budget = df.budget.div(1000000)
```

```python
df.info()
```

```python
df.revenue.value_counts(dropna = False)
```

```python
df.revenue = df.revenue.replace(0, np.nan)
df.revenue = df.revenue.div(1000000)
```

```python
df.rename(columns = {"revenue": "revenue_musd", "budget": "budget_musd"}, inplace = True)
```
